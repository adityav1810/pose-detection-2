```
import numpy as np
import json 
import os
import cv2
from google.colab.patches import cv2_imshow
json_data='/content/drive/MyDrive/keypoints_from_videos_json/stepup_knee/Step - Up - Knee_000000000000_keypoints.json'
```

```
import json 
  
# Opening JSON file 
f = open(json_data,) 
  
# returns JSON object as  
# a dictionary 
data = json.load(f) 
image=cv2.imread('/content/drive/MyDrive/frames/stepupknee/completesquat.jpg')
# Iterating through the json 
# lisf
keypoints=list(data.items())[1][1][0]['pose_keypoints_2d']
for keys in range(0,len(keypoints),3):
  image = cv2.circle(image, (int(keypoints[keys]),int(keypoints[keys+1])), radius=10, color=(0, 0, 255), thickness=-1)
cv2_imshow(image)  

  
# print(list(d.items())[0])
    
  
# Closing file 
f.close() 
```

```

json_dir='/content/drive/MyDrive/posedetection/output_json_coco_final/'
final_data=np.array([])
for filename in os.listdir(json_dir):
  f = open(json_dir+filename,)
  data = json.load(f)
  keypoints=list(data.items())[1][1][0]['pose_keypoints_2d']
  image_keypoints=[]
  
  for keys in range(0,len(keypoints),3):
    image_keypoints=np.append(image_keypoints,np.array([keypoints[keys],keypoints[keys+1]]))
  final_data=np.append(final_data,image_keypoints)
  print(final_data.shape)

  f.close()
final_data=final_data.reshape(15,36)
```
